**Data Cleaning**

```python
# Check missing values
df.isnull()
```

```python
# Check missing values
df.isnull().sum()
```

```python
# mean() calculates average value of the column
# fillna() replaces missing values with given value

df['Final Exam Marks (out of 100)'] = df['Final Exam Marks (out of 100)'].fillna(df['Final Exam Marks (out of 100)'].mean())
# we use this because Mean keeps data statistically balanced
```

```python
# mode() returns most frequent value
# [0] is used to get the actual value from mode result

df['Result'] = df['Result'].fillna(df['Result'].mode()[0])

# we use this Mode preserves most common category
```

```python
# duplicated() finds duplicate rows
# drop_duplicates() removes them

df = df.drop_duplicates()
#Duplicate data causes model bias
```

```python
# astype() converts column data type
# int ensures Marks are integer values

df['Marks'] = df['Marks'].astype(int)
#ML models expect correct data types

#Prevents calculation errors
```

```python
# drop() removes unnecessary columns
# axis=1 means column-wise removal

df = df.drop('Student_ID', axis=1)
#ID columns do not help prediction
```

**Train-Test Split**

Training data  to train the classification model

Testing data  to evaluate how well the model performs on unseen data

```python
# X contains input features used for prediction
# y contains target labels (classification output)

X = df[['Daily Study Hours', 'Final Exam Marks (out of 100)']]   # Features
y = df['Result']                   # Target
print(X)
print(y)
```

```python
# Import train_test_split from sklearn
# This function is used to divide data into train and test sets
from sklearn.model_selection import train_test_split
# Split the data into training and testing sets

X_train, X_test, y_train, y_test = train_test_split(
    X,              # Feature data
    y,              # Target labels
    test_size=0.2,  # 20% data for testing
    random_state=42,# Ensures same split every time
    stratify=y      # Maintains class balance in both sets
)


```

```python
# Print shapes of train and test sets
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

```python
# Import Logistic Regression model
from sklearn.linear_model import LogisticRegression
# Create model
model = LogisticRegression()
# Train model using training data
model.fit(X_train, y_train)
# Evaluate model using test data
model.score(X_test, y_test)
```

**Overfitting vs Underfitting**

```python
# Calculate accuracy on training data
# model.score() returns accuracy of the model

train_accuracy = model.score(X_train, y_train)

# Calculate accuracy on testing data
# This shows how model performs on unseen data

test_accuracy = model.score(X_test, y_test)
# Display both accuracies
train_accuracy, test_accuracy
# To compare training vs testing performance
```

```python
#Identifying Overfitting
# Example scenario (conceptual):
# Training accuracy = 98%
# Testing accuracy = 70%

# This means model memorized training data
# But failed to generalize
```

```python
#Identifying Underfitting
# Example scenario (conceptual):
# Training accuracy = 60%
# Testing accuracy = 58%

# Model failed to learn even training patterns
```

```python
# Calculate difference between training and testing accuracy

accuracy_gap = train_accuracy - test_accuracy
accuracy_gap
```

**Regularization (L1 & L2)**

Regularization is a technique used to prevent overfitting

```python
# Import Logistic Regression model
from sklearn.linear_model import LogisticRegression

# Create Logistic Regression model with L2 regularization
# penalty='l2' applies Ridge regularization
# This is the default regularization in Logistic Regression

model_l2 = LogisticRegression(penalty='l2')

# Train the model using training data
model_l2.fit(X_train, y_train)

# Evaluate model performance on test data
model_l2.score(X_test, y_test)
```

```python
# Create Logistic Regression model with L1 regularization
# penalty='l1' applies Lasso regularization
# solver='liblinear' is required for L1

model_l1 = LogisticRegression(penalty='l1', solver='liblinear')

# Train the model
model_l1.fit(X_train, y_train)

# Evaluate model
model_l1.score(X_test, y_test)

```

**Hyperparameter Tuning (GridSearchCV)**

finding the best combination of model parameters

```python
# Import GridSearchCV from sklearn
# GridSearchCV automatically tests multiple parameter combinations

from sklearn.model_selection import GridSearchCV

# Define hyperparameters to test
# C  regularization strength
# penalty  type of regularization
# solver  optimization algorithm

param_grid = {
    'C': [0.01, 0.1, 1, 10],
    'penalty': ['l1', 'l2'],
    'solver': ['liblinear']
}
```

```python
# Create GridSearchCV object
# LogisticRegression()  base classification model
# cv=5  5-fold cross validation
# scoring='accuracy'  evaluation metric

grid = GridSearchCV(
    LogisticRegression(),
    param_grid,
    cv=5,
    scoring='accuracy'
)
```

```python
# Fit GridSearch on training data
# GridSearch trains model multiple times automatically

grid.fit(X_train, y_train)
```

```python
# Extract best hyperparameter values

grid.best_params_
```

```python
# Get best trained model from GridSearch

best_model = grid.best_estimator_
```

```python
# Evaluate optimized model on test data

best_model.score(X_test, y_test)
```

```python
#When to Use GridSearchCV?
#Small to medium datasets
#When model performance is critical
#Before final model selection
```

**ML Pipeline & Data Leakage Prevention**

combine multiple machine learning steps
(like scaling, SMOTE, model training)

```python
# Import Pipeline to create ML workflow
from sklearn.pipeline import Pipeline

# Import StandardScaler for feature scaling
from sklearn.preprocessing import StandardScaler

# Import Logistic Regression for classification
from sklearn.linear_model import LogisticRegression

# WHY we import these:
# Pipeline  to connect steps
# StandardScaler  to normalize data
# LogisticRegression  classification model
```

```python
# Create pipeline with ordered steps
# Each step is executed in sequence

pipeline = Pipeline([
    ('scaler', StandardScaler()),        # Step 1: Feature Scaling
    ('model', LogisticRegression())      # Step 2: Classification
])
```

```python
# Train the entire pipeline
# fit() applies scaling ONLY on training data

pipeline.fit(X_train, y_train)
```

```python
# Evaluate pipeline performance on test data

pipeline.score(X_test, y_test)
```

```python
# Predict class labels using pipeline

y_pred = pipeline.predict(X_test)
y_pred

# WHY we use pipeline.predict():
# Automatically applies scaling
# Then applies trained classification model
```

```python
# CORRECT APPROACH

pipeline.fit(X_train, y_train)
pipeline.score(X_test, y_test)

# WHY this is correct:
# Training data only used for learning
# Test data remains untouched
```

```python
# Import Pipeline
from sklearn.pipeline import Pipeline

# Create pipeline
pipeline = Pipeline([
    ('scaler', StandardScaler()),           # Scaling
    ('model', LogisticRegression())          # Classification
])

# Train pipeline
pipeline.fit(X_train, y_train)

# Predict
pipeline.score(X_test, y_test)
```
